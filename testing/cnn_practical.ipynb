**Formula for output size of Convolutional Neural Network Layer** : $\frac{N - F + 2P}{S} + 1$ 

```python
import numpy as np
```

```python
X = np.random.randint(0, 256, (6000, 32,32,1))
Y = np.random.randint(0,6, 6000)
```

```python
X = X.astype('float') / 255
```

```python
from tensorflow.keras.utils import to_categorical
Y_onehot = np.zeros(shape=(6000, 6))
```

```python
for i in range(Y.size):
    Y_onehot[i][Y[i]] = 1
```

```python
Y_onehot
```

```python
X.shape
```

```python
Y_onehot.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Dense, Input, MaxPooling2D, Flatten
from tensorflow.keras.optimizers import SGD
```

```python
cnn_model = Sequential([
    # input layer
    Input(shape=(32,32,1)),
    # layer 1
    Conv2D(filters=16, kernel_size=(3,3), activation='relu'),
    # Max Pooling Layer 1
    MaxPooling2D(pool_size=(2,2), strides=2),
    # layer 2
    Conv2D(filters=32, kernel_size=(3,3), strides=1),
    # Max Pooling Layer 2
    MaxPooling2D(pool_size=(2,2), strides=2),
    # Flatten
    Flatten(),
    Dense(6, activation="softmax")
])
```

```python
cnn_model.compile(optimizer=SGD(learning_rate=0.005), metrics=["accuracy"], loss="categorical_crossentropy")
```

```python
history_cnn = cnn_model.fit(X, Y_onehot, batch_size=6000, epochs=10, verbose=1)
```
